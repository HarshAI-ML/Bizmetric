12. Ask user to enter information about salary of the employee per year and rating received as A, B, C, D

If salary upto 5 lpa then increament based on ratings are A = 16% , B=12%, C= 10%, D=6%

If salary upto 23 lpa then increament based on ratings are A = 7% , B=5%, C= 4%, D=no

```python
def salary():
      currsal=input("current sal")
      currrat=input("currrent rating recieved").upper()
      if currrat not in ['A', 'B', 'C', 'D']:
        return "Invalid rating"

      try:
        currsal=int(currsal)
      except ValueError:
        return "Salary must be numeric"

      if currsal>1500000:
        if currrat=='A':
          updatedsal=currsal*108/100
        elif currrat=='B':
          updatedsal=currsal*106/100
        elif currrat=='C':
          updatedsal=currsal*104/100
        elif currrat=='D':
          updatedsal=currsal*100/100
        else:
          return("Input correct rating")

      elif currsal>1000000:
        if currrat=='A':
          updatedsal=currsal*114/100
        elif currrat=='B':
          updatedsal=currsal*110/100
        elif currrat=='C':
          updatedsal=currsal*108/100
        elif currrat=='D':
          updatedsal=currsal*106/100
        else:
          return("Input correct rating")

      elif currsal>500000:
        if currrat=='A':
          updatedsal=currsal*116/100
        elif currrat=='B':
          updatedsal=currsal*112/100
        elif currrat=='C':
          updatedsal=currsal*110/100
        elif currrat=='D':
          updatedsal=currsal*106/100

      else:
        if currrat=='A':
          updatedsal=currsal*107/100
        elif currrat=='B':
          updatedsal=currsal*105/100
        elif currrat=='C':
          updatedsal=currsal*104/100
        elif currrat=='D':
          updatedsal=currsal*100/100
        else:
          return("Input correct rating")

      return updatedsal
```

14. Digitalize the book allotment process for school. Charges are mentioned here in the given table:

```python
std_map = {
    1: '1to4',
    2: '5to8',
    3: '9to10'
}
bookcost = 0
notebookcost = 0
totalcost=0


books = {
    '1to4': {'hindi':60,'marathi':60,'english':60,'science':70,'maths':80},
    '5to8': {'hindi':60,'marathi':60,'english':60,'science':70,'maths':80},
    '9to10': {'hindi':60,'marathi':60,'english':60,'science':70,'maths':80}
}

notebooks = {
    '100pages': {'squares':40,'4lines':30,'2lines':30,'singleline':60,'a4':100},
    '200pages': {'squares':40,'4lines':30,'2lines':30,'singleline':60,'a4':100}
}

while(True):
    if totalcost>0:
      exit=int(input("press 1 to exit 2 to continue "))
      if exit==1:
        break
      elif exit==2:
        pass
      else:
        print("wrong input .Have a good day ")
        break
    booktype = input('enter books or notebooks: ').lower()

    if booktype == 'books':
        std = int(input('press 1 for 1to4, 2 for 5to8, 3 for 9to10: '))
        subject = input('enter subject: ').lower()
        bookquantity = int(input('enter book quantity: '))

        std_key = std_map.get(std)

        if std_key and subject in books[std_key]:
            bookcost = books[std_key][subject] * bookquantity
        else:
            print("Invalid standard or subject")

    elif booktype == 'notebooks':
        notebooktype = input('enter 100pages or 200pages: ').lower()
        notebooklinetype = input('enter notebook line type: ').lower()
        notebookquantity = int(input('enter notebook quantity: '))

        if notebooktype in notebooks and notebooklinetype in notebooks[notebooktype]:
            notebookcost = notebooks[notebooktype][notebooklinetype] * notebookquantity
        else:
            print("Invalid notebook type or line type")

    else:
        print("Invalid input")

    totalcost = totalcost+bookcost + notebookcost
    print("Total Cost:", totalcost)
```

```python
class Book_allotment:
   def __init__(self):

      # self.std_map = {
      #     1: '1to4',
      #     2: '5to8',
      #     3: '9to10'
      # }
      self.bookcost = 0
      self.notebookcost = 0
      self.totalcost=0

   @classmethod
   def initialization(cls):
      cls.std_map = {
          1: '1to4',
          2: '5to8',
          3: '9to10'
      }

      cls.books = {
          '1to4': {'hindi':60,'marathi':60,'english':60,'science':70,'maths':80},
          '5to8': {'hindi':60,'marathi':60,'english':60,'science':70,'maths':80},
          '9to10': {'hindi':60,'marathi':60,'english':60,'science':70,'maths':80}
      }

      cls.notebooks = {
          '100pages': {'squares':40,'4lines':30,'2lines':30,'singleline':60,'a4':100},
          '200pages': {'squares':40,'4lines':30,'2lines':30,'singleline':60,'a4':100}
      }
   def order(self):
      while(True):
          if self.totalcost>0:
            exit=int(input("press 1 to exit 2 to continue "))
            if exit==1:
              break
            elif exit==2:
              pass
            else:
              print("wrong input .Have a good day ")
              break
          self.booktype = input('enter books or notebooks: ').lower()

          self.bookcost = 0
          self.notebookcost = 0

          if self.booktype == 'books':
              std = int(input('press 1 for 1to4, 2 for 5to8, 3 for 9to10: '))
              subject = input('enter subject: ').lower()
              self.bookquantity = int(input('enter book quantity: '))

              std_key = self.std_map.get(std)

              if std_key and subject in self.books[std_key]:
                  self.bookcost = self.books[std_key][subject] * self.bookquantity
              else:
                  print("Invalid standard or subject")

          elif self.booktype == 'notebooks':
              self.notebooktype = input('enter 100pages or 200pages: ').lower()
              self.notebooklinetype = input('enter notebook line type: ').lower()
              self.notebookquantity = int(input('enter notebook quantity: '))

              if self.notebooktype in self.notebooks and self.notebooklinetype in self.notebooks[self.notebooktype]:
                  self.notebookcost = self.notebooks[self.notebooktype][self.notebooklinetype] * self.notebookquantity
              else:
                  print("Invalid notebook type or line type")

          else:
              print("Invalid input")

          self.totalcost = self.totalcost+self.bookcost + self.notebookcost
          print("Total Cost:", self.totalcost)
```

35. List2 = [2,70,'work', para, 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 3,10,302.5]

```python
List2 = [2, 70, 'work', 'para', 2.5,
         [1, 2, 3],
         (1, 2),
         {1, 2},
         {1: 'a', 2: 'b'},
         3, 10, 302.5]

result = []

for item in List2:
    if isinstance(item, (int, float, str, list, tuple)):
        result.append(item * 2)
    elif isinstance(item, set):
        result.append({x * 2 for x in item})
    elif isinstance(item, dict):
        result.append({k * 2: v for k, v in item.items()})

print(result)
```

```python
def bill_layout():
    total=0
    print("|" + "-"*60+ "|")
    print("|{:^60}|".format("Welcome to Hotel Shinde's"))
    print("|"+"-" * 60+"|")
    print("|{:<30}{:>30}|".format("Order id :", "Receipt"))
    print("|"+"-" * 60+"|")
    print("|{:<10}{:<21}{:>10}{:>19}|".format("sr","Menu","Quant","Price"))
    # a=len(order_bill)
    menu=['Idli','sandwhich']
    price=[190,100]
    for i in range(1,len(menu)+1):
      print("|{:<6}{:<25}{:>10}{:>19}|".format(i,menu[i-1],"1",price[i-1]))
      total=total+price[i-1]

    print("|"+"-" * 60+"|")
    print("|{:>15}{:>45}|".format("Total",total))
    print("|"+"-" * 60+"|")
```
