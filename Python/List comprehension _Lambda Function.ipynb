Use lambda with map() to square every number in a list

```python
li = [1,2,3,4,5,6]
squares = lambda lst: map(lambda x: x**2, lst)
print(list(squares(li)))
```

Use lambda with filter() to get only odd numbers from a list.


```python
li = [1,2,3,4,5,6]
odd = lambda st: filter(lambda x: x%2!=0, st)
print(list(odd(li)))
```

Use sorted() + lambda to sort a list of tuples by second value.

```python
data = [(1, 5), (2, 2), (3, 8), (4, 1)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

Write a lambda to remove duplicates from a list using filter + set.

Write a lambda to check if two strings are anagrams

```python
is_anagram = lambda s1, s2: sorted(s1.replace(" ", "").lower()) == sorted(s2.replace(" ", "").lower())

print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
```
