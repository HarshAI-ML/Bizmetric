```python
class DegreeCost:
    def __init__(self, subject, analytics, hostel, food_months, transport):
        self.subject = subject
        self.analytics = analytics.upper()
        self.hostel = hostel.upper()
        self.food_months = food_months
        self.transport = transport.lower()

        # Fees
        self.core_fee = 200000
        self.analytics_fee = self.core_fee * 1.10
        self.hostel_fee = 200000
        self.food_fee_per_month = 2000
        self.transport_fee_semester = 13000
        self.transport_fee_annual = 26000

        self.total_cost = 0

    def calculate_total(self):
        # Subject + Analytics
        if self.subject in ["HR", "Finance", "Marketing"]:
            if self.analytics == "Y":
                self.total_cost += self.analytics_fee
            else:
                self.total_cost += self.core_fee
        elif self.subject == "DS":
            self.total_cost += self.core_fee
        else:
            raise ValueError("Invalid subject entered!")

        # Hostel
        if self.hostel == "Y":
            self.total_cost += self.hostel_fee

        # Food
        self.total_cost += self.food_months * self.food_fee_per_month

        # Transportation
        if self.transport == "semester":
            self.total_cost += self.transport_fee_semester
        elif self.transport == "annual":
            self.total_cost += self.transport_fee_annual
        else:
            raise ValueError("Invalid transport choice!")

        return self.total_cost




subject = input("Enter subject (HR/Finance/Marketing/DS): ").strip()
analytics = input("Opt for Analytics (Y/N): ").strip()
hostel = input("Opt for Hostel (Y/N): ").strip()
food_months = int(input("Food (how many months): "))
transport = input("Transportation (semester/annual): ").strip()

student = DegreeCost(subject, analytics, hostel, food_months, transport)
total=student.calculate_total()
print("Total degree cost is :", total)

```
